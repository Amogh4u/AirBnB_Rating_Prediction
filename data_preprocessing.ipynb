# location using latitude and longitude

```python
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent="geoapiExercises")
```

```python
merged_data['combined_location'] = merged_data["latitude"].astype(str) + "," + merged_data["longitude"].astype(str)
```

```python
def get_location(location):
    lat, lng = location.split(',')
    loc = str(geolocator.geocode(str(lat)+","+str(lng)))
    loc = loc.replace('', '(')
    loc = loc.replace('', ')')
    return geolocator.geocode(str(lat)+","+str(lng))
```

```python
merged_data['location_from_lat_lng'] = merged_data["combined_location"].apply(get_location)
```

```python
# merged_data['location_from_lat_lng'] = merged_data['location_from_lat_lng'].astype(str)
# merged_data['location_from_lat_lng'] = merged_data['location_from_lat_lng'].replace('(', '')
# merged_data['location_from_lat_lng'] = merged_data['location_from_lat_lng'].replace(')', '')
merged_data['location_from_lat_lng'] = merged_data['location_from_lat_lng'].astype(str)
merged_data['location_from_lat_lng']
```

# divide host _verifications  as done for ammenities

```python
merged_data['host_verifications'] = merged_data['host_verifications'].astype(str)
merged_data.host_verifications = merged_data.host_verifications.str.replace(r"{[|]}|\"","").str.strip().str.split(",")
host_verifications_df = pd.DataFrame([vrf for vrf in merged_data.host_verifications.apply(lambda item: dict(map(lambda vrf: (vrf,1),item))).values]).fillna(0)
```

```python
host_verifications_df.columns = host_verifications_df.columns.str.replace('[', '')
host_verifications_df.columns = host_verifications_df.columns.str.replace(']', '')
host_verifications_df.columns = host_verifications_df.columns.str.replace('\\', '__')
host_verifications_df.columns = host_verifications_df.columns.str.replace('__u', '')
host_verifications_df.columns = host_verifications_df.columns.str.replace('\'', '')
host_verifications_df.columns = host_verifications_df.columns.str.replace(' ', '')
host_verifications_df
```

```python
host_verifications_df = host_verifications_df.groupby(host_verifications_df.columns, axis=1).agg(np.max)
host_verifications_df = host_verifications_df.drop(host_verifications_df.columns[0], axis=1)
host_verifications_df
```

```python
merged_data = merged_data.join(host_verifications_df)
```

```python
merged_data = merged_data.drop(columns=["host_verifications"])
```
