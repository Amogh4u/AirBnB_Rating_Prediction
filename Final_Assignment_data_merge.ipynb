```python
import pandas as pd
```

```python
df_listings = pd.read_csv('data/listings.csv')
df_reviews = pd.read_csv('reviews_updated.csv')
```

```python
df_listings = df_listings.rename(columns = {'id': 'listing_id'})
```

```python
df_merged = pd.merge(df_listings, df_reviews, how='inner', on=['listing_id'])
```

```python
df_merged.to_csv('merged_data.csv',index=False)
```
